## Collection

```python
import os
import uuid
import shutil

input_dir = "data/collection"
output_dir = "data/named/images"

# Loop through the files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg"):
        # Construct the new filename by appending '_new' to the original filename
        new_filename = filename.replace(filename, f'{str(uuid.uuid1())}.jpg')
        # Copy the file to the output directory and rename it
        shutil.copy(os.path.join(input_dir, filename), os.path.join(output_dir, new_filename))
```

count file number

```python
# folder path
dir_path = "data/collection"
count = 0
# Iterate directory
for path in os.listdir(dir_path):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        count += 1
print('File count:', count)
```
